#### Covariance and correlation are both used in statistics to understand the relationship between two variables, but they have some key differences:

Covariance:

Measures the direction and magnitude of the linear relationship between two variables.
A positive covariance indicates that the variables tend to move in the same direction. If one variable increases, the other tends to increase as well, and vice versa.
A negative covariance indicates that the variables tend to move in opposite directions. If one variable increases, the other tends to decrease.
Covariance is not standardized, so the value depends on the units of the variables being measured.
It can be difficult to interpret the strength of the relationship from the covariance value alone.
Correlation:

Measures the strength and direction of the linear relationship between two variables.
Similar to covariance, a positive correlation indicates a positive relationship and a negative correlation indicates an opposite relationship.
Correlation is a standardized value, typically ranging from -1 to +1.
A correlation of 1 indicates a perfect positive linear relationship.
A correlation of -1 indicates a perfect negative linear relationship.
A correlation of 0 indicates no linear relationship.
Because it's standardized, correlation is easier to interpret than covariance. It allows you to compare the strength of relationships between different variables regardless of their units.
Here's an analogy:

Imagine two people walking on a tightrope.
Covariance tells you if they are both leaning in the same direction (positive) or opposite directions (negative), and how far they are leaning (magnitude).
Correlation tells you how well they are keeping their balance (strength), regardless of how high the tightrope is (units).
Covariance is a building block for correlation. Correlation can be thought of as a scaled version of covariance, taking the units of the variables into account.

In summary, both covariance and correlation help us understand how two variables are related. Covariance tells you the direction and magnitude of the relationship, while correlation tells you the strength and direction on a standardized scale.

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame()
```

```python
from re import X
x = pd.Series([12,25,68,42,113])
y = pd.Series([11,29,58,121,100])
```

```python
df['x'] = x
df['y'] = y
```

```python
df
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 3))

#Plot scatterplots on each axes
ax1.scatter(df['x'], df['y'])
ax2.scatter(df['x']*2, df['y']*2)

ax1.set_title("Covariance -" + str(np.cov(df['x'],df['y'])[0,1]))
ax2.set_title("Covariance -" + str(np.cov(df['x']*2,df['y']*2)[0,1]))
```

```python
#Correlation

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 3))

# Plot scatterplots on each axes
ax1.scatter(df['x'], df['x'])
ax2.scatter(df['x'], df['y'])

ax1.set_title("Correlation - " + str(df['x'].corr(df['x'])))
ax2.set_title("Correlation - " + str((df['x']).corr(df['y'])))
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 3))

# Plot scatterplots on each axes
ax1.scatter(df['x'], df['y'])
ax2.scatter(df['x']*2, df['y'])

ax1.set_title("Correlation - " + str(df['x'].corr(df['y'])))
ax2.set_title("Correlation - " + str((df['x']*2).corr(df['y']*2)))
```
